#Task 10

```python
s = input("Enter two strings separated by a comma and a space: ")
s1 = ""
s2 = ""
new = ""

for i in range(len(s)):
  if s[i] != ",":
    s1 += s[i]
  else:
    s2 += s[i+2:]
    break

if len(s1) > len(s2):
  for i in range(len(s2)):
    new += s1[i]
    new += s2[i]
  new += s1[len(s2):]
else:
  for i in range(len(s1)):
    new += s1[i]
    new += s2[i]
  new += s2[len(s1):]

print(new)
```
