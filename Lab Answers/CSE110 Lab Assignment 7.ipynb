#Task 7

```python
def show_palindrome(n):
  x=''
  for i in range(1,n+1):
    x+=str(i)
  for j in range(n-1,0,-1):
    x+=str(j)
  return x
print(show_palindrome(5))
```

```python
def remove(l1):
  l2=[]
  for i in l1:
    if i not in l2:
      l2.append(i)
      c=1
    elif i in l2 and c==1:
      l2.append(i)
      c=2
  return l2
l=[1, 2, 3, 3, 3, 3, 4, 5, 8, 8]
print('Removed:',len(l)-len(remove(l)))
print(remove(l))
```

```python
a=int(input('Starting Number: '))
b=int(input('Ending Number: '))
c=int(input('First divisor: '))
d=int(input('Second divisor: '))
def complex(up,low,fd,sd):
  sum=0
  for i in range(up,low):
    if i%(fd*sd)!=0:
      if i%fd==0:
        sum+=i
      elif i%sd==0:
        sum+=i
  return sum
print(complex(a,b,c,d))
```

```python
def alphabet(s):
  x=''
  n=''
  for i in s:
    if i!=' ':
      x+=i
  for i in x:
    if i not in n:
      n+=i
  sum=0
  for i in n:
    if ord('a')<=ord(i)<=ord('j'):
      sum+=1
  if sum==10:
    return 5
  else:
    return 6
string=input('String: ').lower()
for i in range(alphabet(string)):
  print('PSG will win the Champions League this season')
```

```python
def individul_bonus_calculation(name,yearly_income,total_goal,bonus):
  if total_goal>30:
    x=total_goal*(bonus/100*yearly_income)+10000
  elif 20<=total_goal<=30:
    x=total_goal*(bonus/100*yearly_income)+5000
  else:
    x=total_goal*(bonus/100*yearly_income)+0
  print(f'{name} earned a bonus of {int(x)} Taka for {total_goal} goals.')
individul_bonus_calculation('Luis', 80000, 25, 10)
```
