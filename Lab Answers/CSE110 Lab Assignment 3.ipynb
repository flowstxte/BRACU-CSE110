#Task 16

```python
#While

q = int(input("Quantity: "))
print("Enter", q, "numbers")
sum = 0
i = 1
minimum = 0
maximum = 0

while i <= q:
    num = int(input())
    sum += num
    if i == 0:
        minimum = num
        maximum = num
    if num < minimum:
        minimum = num
    if num > maximum:
        maximum = num
    i += 1

print("Maximum", maximum)
print("Minimum", minimum)
print("Average is", sum / q)
```

#Task 28 (Ungraded)

```python
n1 = int(input("Starting Number: "))
n2 = int(input("Ending Number: "))
print(f"Between {n1} and {n2},")
primes=[]
perfects=[]
for num in range(n1,n2+1):
  prime=0
  for i in range(1,num+1):
    if num%i==0:
      prime+=i
      if prime==num+1:
        primes.append(num)
for num in range(n1,n2+1):
  perfect=0
  for i in range(1,num):
    if num%i==0:
      perfect+=i
      if perfect==num:
        perfects.append(num)
if len(primes) == 0 or len(primes) == 1:
  print("Found", len(primes), "prime number")
else:
  print("Found", len(primes), "prime numbers")
if len(perfects) == 0 or len(perfects) == 1:
  print("Found", len(perfects), "perfect number")
else:
  print("Found", len(perfects), "perfect numbers")
if len(primes) == 0 or len(primes) == 1:
  print("Prime number:", ", ".join(str(item) for item in primes))
else:
  print("Prime numbers:", ", ".join(str(item) for item in primes))
if len(perfects) == 0 or len(perfects) == 1:
  print("Perfect number:", ", ".join(str(item) for item in perfects))
else:
  print("Perfect numbers:", ", ".join(str(item) for item in perfects))
```
